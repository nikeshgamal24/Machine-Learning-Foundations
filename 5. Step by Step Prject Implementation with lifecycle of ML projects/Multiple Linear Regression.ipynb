## Multiple Linear Regression

```python
from sklearn.datasets import fetch_california_housing
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
california  = fetch_california_housing()
```

```python
california
```

```python
print(california.data)
```

```python
print(california.target)
```

```python
california.feature_names
```

```python
## Let's prepare the dataframe   -----------------This is the independent feautre 
dataset = pd.DataFrame(california.data,columns=california.feature_names)
dataset.head()
```

```python
## This is the dependent feature
dataset['Price'] = california.target
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.head()
```

## Extract the independent and dependent feature from the dataset 

```python
## Extract the independent and dependent feature from the dataset 
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

## Train Test Split the dataset 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=10)
```

```python
X_train.shape,y_train.shape
```

## Standardize the datapoints

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
type(y_train)
```

```python
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
X_train
```

```python
## Model Training
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train,y_train)
```

```python
## coefficient or slope
regression.coef_  ## for every feature there are slope for each 
```

```python
## intercepts
regression.intercept_
```

```python
y_pred = regression.predict(X_test)
```

```python
y_pred
```

## Performance metrics

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

## R square 
Formula

**R^2 = 1 - SSR/SST**


R^2	=	coefficient of determination
SSR	=	sum of squares of residuals
SST	=	total sum of squares

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,y_pred)
```

```python
score
```

## Adjusted R Square

**Adjusted R2 = 1 – [(1-R2)*(n-1)/(n-k-1)]**

where:

R2: The R2 of the model
n: The number of observations
k: The number of predictor variables

```python
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
